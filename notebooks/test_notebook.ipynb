```python
from importlib import reload
import nfs_ab_tool.utils.dataset as datasets
import nfs_ab_tool.utils.plots as plot

datasets = reload(datasets)
plot = reload(plot)
```

```python
binary_metrics = ['convertion_1', 'convertion_2']
continuous_metrics = ['cont_metric_1', 'cont_metric_2']

metrics = {
    'continuous': continuous_metrics,
    'binary': binary_metrics
}

dataset = datasets.Dataset().generate(
    binary_metrics=binary_metrics,
    continuous_metrics=continuous_metrics,
    group_num=2,
    date_num=14
)
```

```python
dataset['date'].min(), dataset['date'].max()
```

```python
plotter = plot.Plotter(
    group_col='group',
    date_col='date',
    user_col='user_id',
    metrics=metrics,
    start_test_date='2023-11-20',
    end_test_date='2023-12-07',
)
```
